```python
df = pd.read_csv('dailysales.csv')
```

```python
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
daily_df = df.resample('D').sum()
train_df = daily_df['sales'].iloc[:1000]
```

```python
X_t = daily_df['sales'][:1100]
split_idx = int(X_t.shape[0] * 0.8)
training = X_t[:split_idx]
test = X_t[split_idx:]
```

```python
X_preds = list()
for t in range(len(test)):
    temp_train = X_t[:len(training) + t]
    model = ARIMA(temp_train, order=(8, 0, 0))  # Ordre du modèle ARIMA, ajustez au besoin
    model_fit = model.fit()
    predictions = model_fit.predict(start=len(temp_train), end=len(temp_train), dynamic=False)
    X_preds = X_preds + [predictions]
```

```python
# Evaluate the model
test = np.array(test)
X_preds = np.array(X_preds).flatten()

r2 = r2_score(test, X_preds)
print(f'R-squared score: {r2}')
```

```python
# Plotting the results
plt.plot(test, color='blue', label='Real Sales')
plt.plot(X_preds, color='black', label='Predicted Sales')
plt.title('Sales Prediction with ARIMA')
plt.xlabel('Time')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import matplotlib.pyplot as plt

# Charger les données (assurez-vous d'avoir des colonnes 'date' et 'sales')
data = pd.read_csv('dailysales.csv')
data['date'] = pd.to_datetime(data['date'])
data = data.set_index('date')['sales'].resample('D').sum()

# Normalisation des données
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(data.values.reshape(-1, 1))

# Préparation des données pour l'entraînement (séquences temporelles)
def create_sequences(data, seq_length):
    sequences = []
    targets = []
    for i in range(len(data) - seq_length):
        seq = data[i:i+seq_length]
        target = data[i+seq_length]
        sequences.append(seq)
        targets.append(target)
    return np.array(sequences), np.array(targets)

seq_length = 9  # ajustez la longueur de la séquence selon vos besoins
X, y = create_sequences(data_scaled, seq_length)

# Divisez les données en ensembles d'entraînement, de test et de validation
train_size = int(len(X) * 0.7)
test_size = int(len(X) * 0.15)

X_train, X_test, y_train, y_test = X[:train_size], X[train_size:train_size+test_size], y[:train_size], y[train_size:train_size+test_size]
X_val, y_val = X[train_size+test_size:], y[train_size+test_size:]

# Création du modèle LSTM
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(seq_length, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

# Entraînement du modèle avec ensemble de validation
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_val, y_val))

# Évaluation du modèle sur l'ensemble de test
loss = model.evaluate(X_test, y_test)
print(f'Loss (MSE) on test data: {loss}')

# Faire des prédictions
predictions = model.predict(X_test)

# Dénormaliser les prédictions
predictions = scaler.inverse_transform(predictions)
y_test = scaler.inverse_transform(y_test.reshape(-1, 1))

# Afficher les résultats
plt.plot(predictions,color='blue', label='Predictions')
plt.plot(y_test,color='black',label='Actual')
plt.legend()
plt.show()

loss = model.evaluate(X_test, y_test)
print(f'Loss (MSE) on test data: {loss}')
```
